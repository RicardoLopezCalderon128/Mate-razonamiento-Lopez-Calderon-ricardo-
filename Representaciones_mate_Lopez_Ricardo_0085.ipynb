Problema 1:

 La suma de un número más 9 es igual a 21Paso 1: Análisis y resoluciónPara encontrar el número desconocido, seguimos estos pasos:Definir la variable: Llamaremos $x$ al número que no conocemos.Plantear la ecuación: La frase "la suma de un número ($x$) más 9 es igual a 21" se escribe así:$$x + 9 = 21$$Despejar la variable: Para dejar a la $x$ sola, el 9 que está sumando pasa al otro lado de la igualdad restando:$$x = 21 - 9$$Resolver: Realizamos la operación.$$x = 12$$Resultado: El número es 12.

```python
#### Problema 1 : encontrar el numero desconocido
# Problema 1: x + 9 = 21
resultado_total = int(input("Ingrese el resultado total: "))
numero_conocido = int(input("Ingrese el número conocido: "))

# Despejamos x: x = 21 - 9
incognita = resultado_total - numero_conocido

print(f"El número buscado es: {incognita}")
```

Problema 2:



 Las edades de Pedro y AnaPaso 1: Análisis y resoluciónEste problema requiere relacionar dos variables, pero podemos expresarlo todo en función de una sola:Definir las variables:Edad de Ana: $x$ (es nuestra base).Edad de Pedro: $2x$ (ya que tiene el doble que Ana).Plantear la ecuación: La suma de ambas edades es 36:$$x + 2x = 36$$Simplificar: Sumamos las $x$:$$3x = 36$$Despejar la variable: El 3 que está multiplicando a la $x$ pasa al otro lado dividiendo:$$x = \frac{36}{3}$$Resolver:$x = 12$ (Edad de Ana)$2x = 2(12) = 24$ (Edad de Pedro)Resultado: Ana tiene 12 años y Pedro tiene 24 años.


```python
# Problema 2: Edad de Ana (x) + Edad de Pedro (2x) = 36
suma_edades = 36

# La ecuación simplificada es 3x = 36, por lo tanto x = 36 / 3
edad_ana = suma_edades / 3
edad_pedro = edad_ana * 2

print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
